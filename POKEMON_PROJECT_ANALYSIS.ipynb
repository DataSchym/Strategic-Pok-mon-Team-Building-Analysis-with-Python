# Question 2: Analyse de Vitesse par GÃ©nÃ©ration 

```python
# Analysez et prÃ©sentez quelle gÃ©nÃ©ration de PokÃ©mon a, en moyenne, les vitesses (Speed) les plus Ã©levÃ©es.
graph1 = pd.pivot_table(df,
                       values = ("Speed"),
                       index = "Generation",
                       aggfunc = "mean").reset_index()
graph1.plot(x="Generation",
           y = "Speed",
           kind = "bar",
           rot = 0)

plt.xlabel("Generation")
plt.ylabel("Speed")
plt.title("Pokemon Speed by Generation")
plt.show()
```

# Question 3: Visualisation 

```python
# Utilisez matplotlib pour crÃ©er un scatter plot comparant les statistiques d'Attaque (Attack) aux statistiques de DÃ©fense (Defense) pour tous les PokÃ©mon de type Feu et Eau.

# On filtre d'abord le type : 'Fire' ou 'Water'
filtered_df = df[(df['Type 1'] == 'Fire') | (df['Type 1'] == 'Water')]

# Maintant on rÃ©alise le graphique
plt.scatter(filtered_df[filtered_df['Type 1'] == 'Fire']['Attack'], filtered_df[filtered_df['Type 1'] == 'Fire']['Defense'], color='red', label='Fire', alpha=0.5)
plt.scatter(filtered_df[filtered_df['Type 1'] == 'Water']['Attack'], filtered_df[filtered_df['Type 1'] == 'Water']['Defense'], color='blue', label='Water', alpha=0.5)

plt.xlabel("Attack")
plt.ylabel("Defense")
plt.title("Attack & Defense by Fire & Water Types")
plt.show()
```

```python
# crÃ©ez un graphique simple montrant la rÃ©partition du nombre de PokÃ©mon par 'Type 1'

# On compte d'abord les pokÃ©mons par type
type_counts = df['Type 1'].value_counts()
type_counts

# On joint 'type_counts' avec 'df'
Tab2 = df.merge(type_counts, on = 'Type 1')

# On dresse le graphique
graph2 = pd.pivot_table(Tab2,
                       values = ("count"),
                       index = "Type 1").reset_index()
graph2.plot(x="Type 1",
           y = "count",
           kind = "bar",
           rot = 90,
           color = 'orange')

plt.xlabel("Type")
plt.ylabel("Amount")
plt.title("Pokemon Amount by Type")
plt.show()
```

# Question 4 : StratÃ©gie d'Ã‰quipe contre une Ã‰quipe SpÃ©cifique 

```python
# Types et faiblesses des adversaires (selon le PDF)
adversaires = {
    'Emboar': ['Water', 'Ground', 'Psychic'],
    'Seismitoad': ['Grass'],
    'Lilligant': ['Fire', 'Flying', 'Ice'],
    'Archeops': ['Water', 'Electric', 'Ice'],
    'Vanilluxe': ['Fire', 'Rock', 'Steel'],
    'Haxorus': ['Ice', 'Dragon', 'Fairy']
}

# On crÃ©e un dictionnaire pour stocker les candidats potentiels
candidats = {faiblesse: [] for faiblesse in set(sum(adversaires.values(), []))}

# On parcourt chaque PokÃ©mon dans df
for _, row in df.iterrows():
    if row['Generation'] <= 5:
        # On vÃ©rifie si le PokÃ©mon est un bon candidat contre au moins un adversaire
        for faiblesse in candidats.keys():
            if row['Type 1'] == faiblesse or row['Type 2'] == faiblesse:
                candidats[faiblesse].append(row['Name'])

# Ã€ ce stade, 'candidats' contient des noms de PokÃ©mon classÃ©s par type efficace contre les adversaires

# On sÃ©lectionne une Ã©quipe Ã©quilibrÃ©e de maniÃ¨re simple
equipe = []
for type_advantageux in candidats:
    equipe.append(candidats[type_advantageux][0]) 
    if len(equipe) == 6:
        break

print("Ã‰quipe proposÃ©e :", equipe)
```
